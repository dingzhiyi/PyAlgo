## Basic sum in recursive
The only advantage: makes your code shorter?

```python
def sum(n):
    if n==1:
        return 1
    return n+sum(n-1)
```

```python
sum(5)
```

## choose m elements from n elements(set)

```python
def pick(number, picked, toPick):
    if toPick==0:
        return print(picked)
    smallest = min(number) if len(picked)==0 else picked[-1]+1
    for i in range(smallest,max(number)+1):
        print("====================")
        picked.append(i)
        pick(number,picked,toPick-1)
        picked.pop()
```

```python
pick([0,1,2,3,4,5,6],[],5)
```
